# 원형차트 랜더링

```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
output_notebook()
```

다중 선그래프 랜더링시 사용했던 데이터

```python
# prepare some data
x = [1, 2, 3, 4, 5]
y1 = [6, 7, 2, 4, 5]
y2 = [2, 3, 4, 5, 6]
y3 = [4, 5, 5, 7, 2]
```

```python
# create a new plot with a title and axis labels
p = figure(title="Multiple glyphs example", x_axis_label="x", y_axis_label="y")
```

마지막 object 에 대해서만 line -> circle 메서드로 변경하여 그려보자. 이 때 line_width 가 아니라 size를 통해 크기를 지정해주는 점이 다름

```python
# add multiple renderers
p.line(x, y1, legend_label="Temp.", color="blue", line_width=2)
p.line(x, y2, legend_label="Rate", color="red", line_width=2)
p.circle(x, y3, legend_label="Objects", color="yellow", size=12)
```

```python
# show the results
show(p)
```

# Rendering bars

```python
# create a new plot with a title and axis labels
p = figure(title="Multiple glyphs example", x_axis_label="x", y_axis_label="y")
```

vbar의 경우, x와 top을 선언해주며, width, bottom 등을 설정해준다.

```python
# add multiple renderers
p.line(x, y1, legend_label="Temp.", color="blue", line_width=2)
p.vbar(x=x, top=y2, legend_label="Rate", width=0.5, bottom=0, color="red")
p.circle(x, y3, legend_label="Objects", color="yellow", size=12)

# show the results
show(p)
```

위 파라미터를 조정해본다면?

```python
# create a new plot with a title and axis labels
p = figure(title="Multiple glyphs example", x_axis_label="x", y_axis_label="y")
```

vbar의 경우, x와 top을 선언해주며, width, bottom 등을 설정해준다.

```python
# add multiple renderers
p.line(x, y1, legend_label="Temp.", color="blue", line_width=2)
p.vbar(x=x, top=y2, legend_label="Rate", width=0.5, bottom=2, color="red")
p.circle(x, y3, legend_label="Objects", color="yellow", size=12)

# show the results
show(p)
```

bottom을 2로 변경하였더니 차트의 xlim 시작점이 0에서 2로 바뀜!
